```python
import os
import shutil
import random

label_dir = "labels_my-project-name_2026-06-17-10-36-55"
classes = ["right", "cant_in", "cant_drive"]

for split in ["train", "val"]:
    os.makedirs(f"yolo_dataset/images/{split}", exist_ok=True)
    os.makedirs(f"yolo_dataset/labels/{split}", exist_ok=True)

for class_name in classes:
    folder = f"saba_frames/{class_name}"
    images = [f for f in os.listdir(folder) if f.endswith(".jpg")]
    
    # 只保留有對應txt的圖片
    images = [f for f in images if os.path.exists(f"{label_dir}/{f.replace('.jpg', '.txt')}")]
    
    random.shuffle(images)
    split_idx = int(len(images) * 0.8)
    
    train_imgs = images[:split_idx]
    val_imgs = images[split_idx:]
    
    for img in train_imgs:
        shutil.copy(f"{folder}/{img}", f"yolo_dataset/images/train/{img}")
        txt = img.replace(".jpg", ".txt")
        shutil.copy(f"{label_dir}/{txt}", f"yolo_dataset/labels/train/{txt}")
    
    for img in val_imgs:
        shutil.copy(f"{folder}/{img}", f"yolo_dataset/images/val/{img}")
        txt = img.replace(".jpg", ".txt")
        shutil.copy(f"{label_dir}/{txt}", f"yolo_dataset/labels/val/{txt}")

print("整理完成！")
```

```python
import cv2
from ultralytics import YOLO
from IPython.display import display, Image
import ipywidgets as widgets

model = YOLO("best.pt")
cap = cv2.VideoCapture(0)  # 改成0試試

img_widget = widgets.Image(format='jpg')
display(img_widget)

while True:
    ret, frame = cap.read()
    if not ret:
        print("讀取失敗")
        break
    
    results = model(frame, conf=0.3)
    annotated = results[0].plot()
    
    _, buffer = cv2.imencode('.jpg', annotated)
    img_widget.value = buffer.tobytes()

cap.release()
```
